# MBA .xml data parsing. All three xml files parsed (Emails.xml, Prospects.xml, Users.xml)

## Import all the packages that are required for parsing

```python
import pandas as pd
import xml.etree.ElementTree as ET
```

## Parse the "Users.xml" file to "users" Pandas data frame

```python
tree = ET.parse("C:\\Users\\Ashlesh B Shetty\\Desktop\\ZZ\\SpringSemester\\ELP\\ProjectWork\\MBA_Data\\Users.xml")
root = tree.getroot()
STAFF_ID = []
NAME_LAST = []
NAME_FIRST = []
GENDER = []
AGE = []
for i in root.findall("DATA_RECORD"):
    STAFF_ID.append( str(i.find('STAFF_ID').text))
    NAME_LAST.append(str(i.find('NAME_LAST').text))
    NAME_FIRST.append(str(i.find('NAME_FIRST').text))
    if i.find("GENDER") is not None:
        GENDER.append(str(i.find('GENDER').text))
    else :
        GENDER.append(str("NA"))
    if i.find("AGE") is not None:
        AGE.append(int(i.find('AGE').text)) 
    else:
        AGE.append(0)

data_dict = {'STAFF_ID': STAFF_ID,
        'NAME_LAST': NAME_LAST,
        'NAME_FIRST': NAME_FIRST,
        'GENDER': GENDER,
        'AGE': AGE}
users = pd.DataFrame.from_dict(data_dict)
users = users[['STAFF_ID','NAME_FIRST', 'NAME_LAST', 'GENDER','AGE']]
users.head()
```

## Parse the "Prospects.xml" file to "prospects" Pandas data frame

```python
tree = ET.parse("C:\\Users\\Ashlesh B Shetty\\Desktop\\ZZ\\SpringSemester\\ELP\\ProjectWork\\MBA_Data\\Prospects.xml")
root = tree.getroot()

PROSPECT_ID=[]
CURRENT_STATUS=[]
APP_COUNT=[]
COUNTRY_CITIZENSHIP=[]
COUNTRY_PERM_RES=[]
VISA_TYPE=[]
INTERNATIONAL=[]
Ethnicity=[]
GENDER=[]
AGE=[]
ISMILITARY=[]
ORIGIN_STATUS=[]
ORIGIN_STARTDATE=[]
ORIGIN_CATEGORY_DESC=[]
ORIGIN_REASON_DESC=[]

for i in root.findall("DATA_RECORD"):
    if i.find("PROSPECT_ID") is not None:
        PROSPECT_ID.append( str(i.find('PROSPECT_ID').text))
    else: PROSPECT_ID.append(str("NA"))
    if i.find("CURRENT_STATUS") is not None:
        CURRENT_STATUS.append( str(i.find('CURRENT_STATUS').text))
    else: CURRENT_STATUS.append(str("NA"))
    if i.find("APP_COUNT") is not None:
        APP_COUNT.append( str(i.find('APP_COUNT').text))
    else: APP_COUNT.append(str("NA"))
    if i.find("COUNTRY_CITIZENSHIP") is not None:
        COUNTRY_CITIZENSHIP.append( str(i.find('COUNTRY_CITIZENSHIP').text))
    else: COUNTRY_CITIZENSHIP.append(str("NA"))
    if i.find("COUNTRY_PERM_RES") is not None:
        COUNTRY_PERM_RES.append( str(i.find('COUNTRY_PERM_RES').text))
    else: COUNTRY_PERM_RES.append(str("NA"))
    if i.find("VISA_TYPE") is not None:
        VISA_TYPE.append( str(i.find('VISA_TYPE').text))
    else: VISA_TYPE.append(str("NA"))
    if i.find("INTERNATIONAL") is not None:
        INTERNATIONAL.append( str(i.find('INTERNATIONAL').text))
    else: INTERNATIONAL.append(str("NA"))
    if i.find("Ethnicity") is not None:
        Ethnicity.append( str(i.find('Ethnicity').text))
    else: Ethnicity.append(str("NA"))
    if i.find("GENDER") is not None:
        GENDER.append( str(i.find('GENDER').text))
    else: GENDER.append(str("NA"))
    if i.find("AGE") is not None:
        AGE.append( str(i.find('AGE').text))
    else: AGE.append(str("NA"))
    if i.find("ISMILITARY") is not None:
        ISMILITARY.append( str(i.find('ISMILITARY').text))
    else: ISMILITARY.append(str("NA"))
    if i.find("ORIGIN_STATUS") is not None:
        ORIGIN_STATUS.append( str(i.find('ORIGIN_STATUS').text))
    else: ORIGIN_STATUS.append(str("NA"))
    if i.find("ORIGIN_STARTDATE") is not None:
        ORIGIN_STARTDATE.append( str(i.find('ORIGIN_STARTDATE').text))
    else: ORIGIN_STARTDATE.append(str("NA"))
    if i.find("ORIGIN_CATEGORY_DESC") is not None:
        ORIGIN_CATEGORY_DESC.append( str(i.find('ORIGIN_CATEGORY_DESC').text))
    else: ORIGIN_CATEGORY_DESC.append(str("NA"))
    if i.find("ORIGIN_REASON_DESC") is not None:
        ORIGIN_REASON_DESC.append( str(i.find('ORIGIN_REASON_DESC').text))
    else: ORIGIN_REASON_DESC.append(str("NA"))


    
data_dict = {'PROSPECT_ID':PROSPECT_ID,
'CURRENT_STATUS':CURRENT_STATUS,
'APP_COUNT':APP_COUNT,
'COUNTRY_CITIZENSHIP':COUNTRY_CITIZENSHIP,
'COUNTRY_PERM_RES':COUNTRY_PERM_RES,
'VISA_TYPE':VISA_TYPE,
'INTERNATIONAL':INTERNATIONAL,
'Ethnicity':Ethnicity,
'GENDER':GENDER,
'AGE':AGE,
'ISMILITARY':ISMILITARY,
'ORIGIN_STATUS':ORIGIN_STATUS,
'ORIGIN_STARTDATE':ORIGIN_STARTDATE,
'ORIGIN_CATEGORY_DESC':ORIGIN_CATEGORY_DESC,
'ORIGIN_REASON_DESC':ORIGIN_REASON_DESC,
}
prospects = pd.DataFrame.from_dict(data_dict)
prospects = prospects[['PROSPECT_ID',
'CURRENT_STATUS',
'APP_COUNT',
'COUNTRY_CITIZENSHIP',
'COUNTRY_PERM_RES',
'VISA_TYPE',
'INTERNATIONAL',
'Ethnicity',
'GENDER',
'AGE',
'ISMILITARY',
'ORIGIN_STATUS',
'ORIGIN_STARTDATE',
'ORIGIN_CATEGORY_DESC',
'ORIGIN_REASON_DESC']]
prospects.head()
```
