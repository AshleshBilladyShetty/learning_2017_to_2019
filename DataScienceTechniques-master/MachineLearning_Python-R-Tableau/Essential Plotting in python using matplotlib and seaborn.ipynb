# Scatter plot : Matplotlib and its limitations so using Seaborn later

**Scatterplot   PROS and CONS in matplot lib**  
CON: Coloring by category in scatterplot of matplotlib is painful loops  
PRO: Area of scatter and Color gradient can be fiven for continuous variables as shown below 

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'], c = iris['petal_length'], s = (iris['petal_width'])**6)
plt.show()
```

**Scatterplot using Seaborn which allows to color wrt to category with minimal pain**

```python
# library & dataset
df = sns.load_dataset('iris')
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False)
plt.legend(loc='lower right')
plt.show()
```

# Barplot in seaborn

```python
iris.head()
iris['sepal_length_bucket'] = iris['sepal_length'] > 5
sns.barplot(x="species", y="sepal_width", hue="sepal_length_bucket", data=iris)
plt.show()
```

```python
sns.barplot(y="species", x="sepal_width", hue="sepal_length_bucket", data=iris)
plt.show()
```

## Correlation plot

```python
corrmatrix = iris.corr()
plt.rcParams['figure.figsize'] = (5,5)
plt.imshow(corrmatrix, cmap = 'RdYlGn')
plt.colorbar()
# plt.title('logins_count highly correlated with count_of_events, \n club_id_count,& quantity_sum')
plt.xticks(np.arange(corrmatrix.shape[0]),corrmatrix.columns, rotation = 90)
plt.yticks(np.arange(corrmatrix.shape[0]),corrmatrix.columns)
plt.show()
```
