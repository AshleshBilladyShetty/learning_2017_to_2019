# Entropy

Claude Shannon Theory: To transfer one bit of information means to reduce the recepients uncertainity by a factor of 2

### Let us understand how the whole entropy equation logic

**Case1** 
* Suppose the weather is completely random there is 50-50 chance of being either sunny or rainy
* If a weather stations says you that it is going to be rainy tomorrow. Then they have actually reduced your uncertainity by a factor of 2. There were 2 likely options now there is just one. So weather channel actually sent you just 1 bit of useful information

**Case2** 
* Suppose the weather has eight possible states all equally likely, each state with probability of 1/8 = 12.5%  
* If a weather stations gives you tomorrows weather they are . Then they have actually reduced your uncertainity by a factor of 8, which is $2^3$. so they sent you 3 bits of information
* One easy way of finding bits of imformation is by taking log of the total  uncertainities to the base of 2.  

I terms of uncertainity
$$ log_2 (8) = 3$$ 
if uncertaininty was in percentage. The probability of a state, which was forecasted by weather department. In our case all state have probability of 1/8 12.5%. Therefore,
$$ -log_2 (1/8) = 3$$
$$ -log_2 (.125) = 3$$

**Case3** 
* What if possibilities are not equally likely. like 75% sunny and 25% rainy  
* If a weather stations gives you tomorrows weather is rainy . Then they have actually reduced your uncertainity by a factor of 4, which is $2^2$. so they sent you 2 bits of information. There fore unceratinity is inverse of probability
* One easy way of finding bits of imformation is by taking log of the total  uncertainities to the base of 2.  

If probability of rainy and sunny is known, bits of information passed if the weather station predicted it was **RAINY** is  
$$ -log_2 (.25) = 2$$ 

If probability of rainy and sunny is known, bits of information passed if the weather station predicted it was **SUNNY**vis  
$$ -log_2 (.75) = .41$$

There for how much information are you going to get from wether station on an average **Entropy**
$$ \text{How much info are you going to get from weather station on an average} = - \sum_i p_i log_2 p_i $$
$$ \text{entropy} = - (.25 *log_2 (.25) + .75 log_2 (.75)) $$
$$ \text{entropy} = .25 * 2 + .75 * 0.41 $$
$$ \text{entropy} = .81 $$

There fore on an average you will get .81 bits of information from weather station daily

#### NOTE: Cross entropy is similar to entropy. Instead of calculating the log of true entropy we calculate the probability of predicted probability. If predicted probability equals the true probability then cross entropy is equal to the entropy. If the distribution differ then the cross entropy will be greater than the entropy by some number of bits. 

# KL Divergence: (Kullback-Leibler Divergence, used in RBM)
Amount by which cross entropy exceeds the actual entropy is called relative entropy. or more commonly they are called KL divergence
