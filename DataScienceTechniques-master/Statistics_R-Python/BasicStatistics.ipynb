### Basic stats Definitions

**Process** :A mechanism that generates data over time e.g., consumers who experience our product and react to it

**Descriptive statistics** refers to techniques to organize and summarize information into a usable form
**Inferential statistics** refers to techniques for drawing conclusions (estimates, predictions, ‚Ä¶) about the process based on sample data.

**Census:** A sample of an entire population of interest  
**Parameter:** Number that refers to some characteristic of a process or population. Typically unknown  
**Statistic:** Number that refers to some characteristic of a sample. Known  
**Measurement:** Value (numerical, textual, date, ‚Ä¶) obtained from each member of the sample. The statistic is a summary of the measurements  

**Simple random sampling** All possible members equally likely to be selected  
**Stratified sampling** Random sample from defined strata

**Nominal data:** Measurements are in distinct categories with no given order; often qualitative  
**Ordinal data:** Measurement categories are ordered  
**Interval data:** Differences between scale values are comparable; always quantitative

### Probability Theory

**Complement** Event that A does not occur 
$$\begin{eqnarray} A^c \\
P(A^c) = 1 - P(A)  \end{eqnarray} $$   


**Mutually Exclusive ** If A occurs then B cannot occur at the same time
$$\begin{eqnarray} P(A \cap B) = 0  \end{eqnarray} $$  

**Conditional ** P(A | B) Occurence of A given B
$$\begin{eqnarray} P(A | B) = \frac{P(A \cap B)}{P(B)}  \end{eqnarray} $$  

**Independent ** occurence of  A doesnot depend on B 
$$\begin{eqnarray} P(A \cap B) = P(A) \times P(B)  \\ 
 P(A | B) = P(A)  \end{eqnarray} $$  


### Rules of probability

**Rule of Subtraction**  
$$\begin{eqnarray} P(A^c) = 1 - P(A)  \end{eqnarray} $$   

**Rule of Multiplication**  
$$\begin{eqnarray} P(A | B) \times P(B)= P(A \cap B)   \end{eqnarray} $$  

**Rule of Addition**  
$$\begin{eqnarray} P(A \cup B) = P(A) + P(B) - P(A \cap B) \end{eqnarray} $$

### Common Discrete Distribution : BINOMIAL DISTRIBUTION

#### Understanding Binomial Distribution : how to find a probility for x successes in n trials where success of each trial is p
Suppose we have to find probaility of 2 successes in 6 trials when the probability of success in each trial is 0.7

1. What are the possible cases of having 2 succeses in 6 trials that is $ ^6C_2 $  
2. What is the probability of each case (0.7)*(0.7)*(0.3)*(0.3)*(0.3)*(0.3) = $(0.7)^2 \times (0.3)^4 $
3. Therefore, probability of 2 successes in 6 trials when probability of success in each trial is 0.7 is equal to =  
$$ ^6C_2 \times  (0.7)^2 \times (0.3)^4 $$


So for above case let us find the Binomial distribution   
P(X = 0) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_0 \times  (0.7)^0 \times (0.3)^6 $  
P(X = 1) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_1 \times  (0.7)^1 \times (0.3)^5 $  
P(X = 2) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_2 \times  (0.7)^2 \times (0.3)^4 $  
P(X = 3) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_3 \times  (0.7)^3 \times (0.3)^3 $  
P(X = 4) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_4 \times  (0.7)^4 \times (0.3)^2 $  
P(X = 5) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_5 \times  (0.7)^5 \times (0.3)^1 $  
P(X = 6) = \binom{n}{x} (p)^x (1-p)^n-x = $ ^6C_6 \times  (0.7)^6 \times (0.3)^0 $  

```python
from scipy.special import comb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
x = 2.0
n = 6.0
p = 0.2
q = 1.0 - p
P = []
for x in np.arange(0,7):
    prob = comb(n,x)* p**x * q**(n-x)
    print('P(X={}) = {}').format(x, prob )
    P.append(prob)

binom_dist = pd.DataFrame( {'CntTrialsPerCase' : np.repeat(n,n+1),
                           'ProbSuccessPerTrial' : np.repeat(p,n+1),
                            'ProbFailurePerTrial' : np.repeat(q,n+1),
                            'CountOfSuccessesInEachCase': np.arange(0,7),
                           'ProbabilityOfEachCase' : P})

binom_dist['ProbabilityOfEachCase'].plot(kind = 'bar')
plt.show()
binom_dist
```

First, by a version of the Law of Large Numbers, $\lim_{n \to \infty}  \bar{X}_n= \mu $, i.e., $ \bar{X}$ will tend to be around Œº. This is one of the properties that makes $ \bar{X}$ a good point estimate of Œº. The Central Limit Theorem takes us to a place where we can answer the question: How good?

‚Ä¢ Stability

## Chi Square Distribution

### ANOVA Assumption 1 : Methods Normality Check : 
* One method: Check for the normality of each group seperately
* Second Method: Check for the normality for all groups one shot simultaneously
  * When needed, we can check the assumption of normality by checking the normality of the ‚Äúresiduals‚Äù of the analysis. Intuitively, the residuals correspond to the values of the response variable for each individual after removing the between-group differences. ### POST ANOVA NEXT STEPS: One Way ANOVA:

**If we find after the ANOVA test that there is relationship then the next question will be which are the 2 groups within the nominal variable that are significantly differentiating our response variable?**

So What we do is now we test every pair of levels within the nominal variable. Test for all possible pairs, But Caution

**BUT CAUTION**  
* For k=4 we will have 6 pair wise test. For k=5 we will have 10 pairwise test. So for increase in k number of tests increases exponentially.
* So for example if I have 5 levels in nominal variables I will have 10 pairwise test. And If i'm okay with 10% error or 90% Confidence for each test in 10 tests. Then the overall error rate will be way higher when we aggregate error of all the pairsise test.As k increases the likelihood of getting a relationship just by random chance increases greater and greater. 
* So to address this it is better to have a threshold for each pair wise test, which we conduct simultaneously, lower thant the overall threshold so that overall we maintain the desired threshold when aggregated. So if we take less than 10% as error threshold for each pairwise test, therefore for overall level we will maintain 10% error threshold. 
* So Tukey's Honest Significant Difference (Tukey's HSD ) test helps us to do that
