# Data Scientist job in California

### This code extracts all the data scientist jobs in California that were sponsored VISA and does EDA on them
Without this code to get to see all the companies you will have to go to 255 next pages but with this code it took just 3 mins to extract complete data and now you can quickly do all you filtering either in pandas or Excel

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate
import numpy as np

df1 = pd.DataFrame()

for i in np.arange(1,255):
    link = "https://www.myvisajobs.com/H1B-Visa/SearchLCA.aspx?JT=data%20scientist&WS=CA&O1=Employer&O2=JobTitle&Y=2017&PN=" + str(i)
    res = requests.get(link)
    soup = BeautifulSoup(res.content,'lxml')
    table = soup.find_all('table')[0] 
    df = pd.read_html(str(table))
    df0 = df[0].iloc[16:,0:6]
    df1 = pd.concat([df1 , df0])
```
