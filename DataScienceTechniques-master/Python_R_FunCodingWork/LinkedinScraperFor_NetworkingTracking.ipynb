Below code extracts the name and headline of each contact

```python
#find number of files in a folder
import os, os.path
DIR = 'C:\\Users\\Ashlesh B Shetty\\Desktop\\files'
filecount = len([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name))])
```

```python
import codecs
import re
import numpy as np

data = codecs.open("C:\\Users\\Ashlesh B Shetty\\Desktop\\files\\download.html", 'r','utf-8').read()
name = [re.search('app-argument=voyager://in/(.*)trk=vw_smart_app_banner', data).group(1)[:-2]]
temp = re.search(',paging&quot;,&quot;\$type&quot;:&quot;com.linkedin.voyager.identity.profile.VolunteerExperienceView&quot;},{&quot;firstName&quot;:&quot;(.*)&quot;,&quot;\$deletedFields&quot;:', data).group(1).split('&quot;')
firstname = [temp[0]]
lastname = [temp[-1]]
headline = [re.search('&quot;headline&quot;:&quot;(.*)&', data).group(1).split('quot;')[0][:-1]]

for i in np.arange(filecount-1):
    data = codecs.open("C:\\Users\\Ashlesh B Shetty\\Desktop\\files\\download (" + str(i+1) +").html", 'r','utf-8').read()
    temp = re.search(',paging&quot;,&quot;\$type&quot;:&quot;com.linkedin.voyager.identity.profile.VolunteerExperienceView&quot;},{&quot;firstName&quot;:&quot;(.*)&quot;,&quot;\$deletedFields&quot;:', data).group(1).split('&quot;')
    firstname.append(temp[0])
    lastname.append(temp[-1])
    name.append(re.search('app-argument=voyager://in/(.*)trk=vw_smart_app_banner', data).group(1)[:-2])
    headline.append(re.search('&quot;headline&quot;:&quot;(.*)&', data).group(1).split('quot;')[0][:-1])
```

```python
name1 = []
url = []
for i in urllist:
    name1.append(re.search('https://www.linkedin.com/in/(.*)/',i).group(1))
    url.append(i[:-1])
```

```python
df1 = pd.DataFrame({'urllist':url,
              'name':name})
```

```python
df2 = pd.DataFrame({'firstname':firstname,
              'lastname':lastname,
              'name':name1,
             'headline':headline})
```

```python
company = "whatsapp"
```

```python
df3 = df1.merge(df2, on = 'name', how = 'left')
df3['full name'] = df3['firstname'] +' '+ df3['lastname']
df3['company'] = company #Manual intervention
df3[['company','headline','full name','urllist',]]
```
