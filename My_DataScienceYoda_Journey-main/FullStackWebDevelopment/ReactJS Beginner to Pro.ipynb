### Spread and rest operators

Spread and rest operators are represented by three dots **...**

Spread: Used to split up array elements OR object properties  
Rest: used to merge a list of function arguments into an array  

```js


// Spread examples with arrays
k = [1,2,3]
L = [...k,8]
console.log(k)
console.log(L)
// [1, 2, 3]
// [1, 2, 3, 8]

// Spread examples with object
k = {'ash':2}
L = {...k,'she':1}
console.log(k)
console.log(L)
// [object Object] {
//   ash: 2
// }
// [object Object] {
//   ash: 2,
//   she: 1
// }

// Rest examples with 
const filter = (...args) => {
  return args.filter(el => el ===  1);
}

console.log(filter(1,1,3));
// [1,1]
```

### Destructuring

Similar to spread but here you can easily etract array elements or object properties and store them in variables

```js
// array destructuring
[a,b] = ['Hello','Hi']
console.log(a)
console.log(b)

//object destructuring
{name} = {name:'ash', d:'she'}
console.log(name)
```

### Primitive type and reference type

```js
// Primitive: Creates a real copy of numbers

const number = 1121;
const number1 = number;
console.log(number1)
// 1121

// Reference type: Object and arrays dont create a copy but add pointers

const person = {ash: '123'};
const person1 = person;
person.ash = '456';
console.log(person1)

// [object Object] {
//   ash: "456"
// }


// How to avoid Reference type:Use spread ... for Object and arrays 

const aeasdf = {ash: '123'};
const xvb = {  ...aeasdf};
aeasdf.ash = '456';
// console.log(xvb)
// [object Object] {
//   ash: "123"
// }
```

### Refreshing array functions

```js
// map functions
const arr1 = [1,2,3];
const arr2 = arr1.map((num) => num*2);
const arr3 = arr1.map((num) => {
  return num*3;
});
console.log(arr1) // [1, 2, 3]
console.log(arr2) // [2, 4, 6]
console.log(arr3) // [3, 6, 9]
```
