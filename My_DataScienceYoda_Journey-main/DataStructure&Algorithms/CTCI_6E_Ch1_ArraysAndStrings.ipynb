**(4)** ord() function gives [unicode](https://www.excelanytime.com/excel/index.php?option=com_content&view=article&id=116:ascii-code-extended-ascii-characters-8-bit-system-and-ansi-code&catid=78&Itemid=474) decimal for a character: ord('A') gives 65. Chr() is opposite of ord: chr(65) gives 'A'.https://www.geeksforgeeks.org/ord-function-python/

#### My solution 1: Complexity   $O(n^{2})$
here the complexity is $O(n^{2})$ because for loop has a replace function inside

```python
%%time
# My first attempt solution

def is_unique(string):
    alpha = 'abcdefghijklmnopqrstuvwxyz'
    alphalen = len(alpha)
    unique_bool = True
    for i in string:
        alpha = alpha.replace(i,'')
        if alphalen == len(alpha):
            unique_bool = False
            break
        else:
            alphalen = len(alpha)
    
    if unique_bool:
        value =  'No duplicate characters'
    else:
        value = 'Has duplicate characters'
    return print(value)
```

```python
print('apple') 
is_unique('apple')

print('\naple') 
is_unique('aple')
```

#### Online solution 1: Complexity $O( n \log(n))$ https://www.geeksforgeeks.org/determine-string-unique-characters/

here the complexity is $O( n \log(n))$ because sorted is $O( n \log(n))$ and then we do for loop which is $O(n)$. there fore overall complexity when additive it takes the one with higher complexity: $O( n \log(n)) + O(n) = O( n \log(n))$

```python
def sorted_unique(string):
    alpha = sorted(string)
    alphalen = len(alpha)
    unique_bool = True
    for i in range(alphalen-1):
        if alpha[i] == alpha[i+1]:
            unique_bool = False
            break
    
    if unique_bool:
        value =  'No duplicate characters'
    else:
        value = 'Has duplicate characters'
    return print(value)
```

```python
print('apple') 
sorted_unique('apple')

print('\naple') 
sorted_unique('aple')
```

### 1.2 Check permutations is anagram
Given two strings, wrtie a method to decide if one is permutation of the other

#### My solution 1: Complexity   $O(n \log(n))$
here the complexity is  $O(n \log(n))$ because sorted function complexity is  $O(n \log(n))$

```python
def anagram(string1, string2):
    unique_bool = False
    
    if sorted(string1) == sorted(string2):
        unique_bool = True
        
    if unique_bool:
        value =  'Strings are anagram or permutations of each other'
    else:
        value = 'Strings are NOT anagram or permutations of each other'
    return print(value)        
```

```python
anagram('ash', 'sha')
anagram('ash', 'shaa')
```

#### My solution 2: Complexity   $O(n)$ using dictionaries
In dictionary insertion of a key is $O(1)$ so for N items its going to be $O(n)$

```python
def anagram_dict(string1, string2):
    unique_bool = False
    s1 = {}
    s2 = {}
    
    for i in string1:
        if i in s1:
            s1[i] += 1
        else:
            s1[i] = 1

    for i in string2:
        if i in s2:
            s2[i] += 1
        else:
            s2[i] = 1

            
    if s1 == s2:
        unique_bool = True
        
    if unique_bool:
        value =  'Strings are anagram or permutations of each other'
    else:
        value = 'Strings are NOT anagram or permutations of each other'
    return print(value)        
```

```python
anagram_dict('ang', 'gnas')
anagram_dict('agn', 'gna')
```

### **1.4 Palindrome Permutation:**   
Given a string, write a function to check if it is a permutation of a palinÂ­ drome. A palindrome is a word or phrase that is the same forwards and backwards. A permutation is a rearrangement of letters. The palindrome does not need to be limited to just dictionary words.  
EXAMPLE  
Input: Tact Coa  
Output: True (permutations: "taco cat", "atco eta", etc.)  

#### My solution 1:  Complexity $ O(n)$

```python
def pali_perm(string):
    dict_char = {}
    for i in string.lower():
        if i == ' ':
            continue
        dict_char[i] = dict_char.get(i,0) + 1
    
    if len([ 1 for i in dict_char if dict_char.get(i) % 2]) > 1 :
        value = 'It is Not a Palindrome Permutation'
    else:
        value = 'It is a Palindrome Permutation'
    
    return value
        
```

```python
pali_perm('Tact Coa')
```

### 1.9 String Rotation:
Assumeyou have a method isSubstringwhich checks if one word is a substring of another. Given two strings, sl and s2, write code to check if s2 is a rotation of sl using only one call to isSubstring (e.g.,"waterbottle" is a rotation of"erbottlewat").

```python
# solution to leetcode https://leetcode.com/problems/rotate-string/
def str_rot(s, goal):
    for i, v in enumerate(s):
        if s == goal:
            return True
        if s[i:] + s[:i] == goal:
            return True
    return False
```

```python
str_rot( '', '')
```
